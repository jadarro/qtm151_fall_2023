````python
# Write your own code
# the data set is the same as the original data set we were working with 
data_sorted = wdi_long.sort_values(["date", "country_name"],ascending=True)
data_sorted
````

(c) Compute within-unit time differences

<font size = "3">

**Example**
- Each country has data for multiple dates

```

            .diff())
```

**Instructions**:

````python
# Write your own code
data_sorted["diff_prop_urban"] = data_sorted.groupby("country_name")["prop_urban"].diff()
data_sorted
````

(d) Plot time series for single country

<font size = "3">

- Use ".query()" to subset "data_sorted" for rows that satisfy

$\quad$  ```  'country_name == "United States"' ```
 
- Produce a line plot with the evolution of "prop_urban" over time in the <br>
United States



````python
# Write your own code
subset = data_sorted.query('country_name == "United States"')
plt.plot("date", "prop_urban",
          data = data_sorted.sort_values("date",ascending = True).query('country_name == "United States"'))
plt.xlabel("date")
plt.ylabel("Proportion Urban")
plt.title("Urban Proportion in United States (2016-2020)")
````
